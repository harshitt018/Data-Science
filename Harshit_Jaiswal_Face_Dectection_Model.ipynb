# **Face Detection**    
**Face model can be used for dectecting single and even multiple face detection**

Methods and parameters specified in the above code:    
**detectMultiScale():** The `'detectMultiScale()'` method is used to identify faces of different sizes in the input image.    
**grey_image:** The first parameter in this method is called grey_image, which is the grayscale image we created previously.   
**scaleFactor:** This parameter is used to scale down the size of the input image to make it easier for the algorithm to detect larger faces. In this
case, we have specified a scale factor of 1.1, indicating that we want to reduce the image size by 10%.   
**minNeighbors:** The cascade classifier applies a sliding window through the image to detect faces in it. You can think of these windows as
rectangles.    
Initially, the classifier will capture a large number of false positives. These are eliminated using the minNeighbors parameter, which specifies
the number of neighboring rectangles that need to be identified for an object to be considered a valid detection.  

To summarize, passing a small value like 0 or 1 to this parameter would result in a high number of false positives, whereas a large number
could lead to losing out on many true positives.   

The trick here is to find a tradeoff that allows us to eliminate false positives while also accurately identifying true positives.   
   
**minSize:** Finally, the minSize parameter sets the minimum size of the object to be detected. The model will ignore faces that are smaller than
the minimum size specified.

# **Import the OpenCV Package and Read the Image**

```python
#Import the OpenCV Package and Read the Image
import cv2
```

# **1. For single face dectection.**

# **Importing the image**

```python
Path = '/content/WhatsApp Image 2024-06-08 at 09.28.04.jpeg'
image = cv2.imread(Path)
```

# **Print the dimensions of this array**

```python
#print the dimensions of this array
image.shape
```

**Notice that this is a 3-dimensional array The array’s values represent the picture’s height, width, and channels respectively. Since this is a color
image, there are three channels used to depict it - blue, green, and red (BGR).**

# **Convert the Image to Grayscale**

```python
# Convert the Image to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

**This array only has two values since the image is grayscale and no longer has the third color channel.**

# **Load the Classifier**

```python
#Load the Classifier
face_clf = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

# **Perform the Face Detection**

```python
#Perform the Face Detection
faceDetected = face_clf.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40,40))
```

# **Drawing a Bounding Box**

```python
#Drawing a Bounding Box
for (x, y, w, h) in faceDetected: cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

# **Displaying the Image**

```python
#Displaying the Image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# **Face Dectecting**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.imshow(image_rgb)
plt.axis('off')
```
